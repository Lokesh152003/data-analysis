```python
from bs4 import BeautifulSoup
import requests as rq
```

```python
url1="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
table=rq.get(url1)
```

```python
code1=BeautifulSoup(table.text,'html')
print(code1)
```

```python
a=code1.find('table')
print(a)
```

```python
head=a.find_all('th')
print(head)
```

```python
h=[i.text for i in head]
print(h)
```

```python
mydf1=pd.DataFrame(columns=h)
print(mydf1)
```

```python
rows=a.find_all('td')
print(rows)
```

```python

x=[i.text for i in rows]
print(x)
```
