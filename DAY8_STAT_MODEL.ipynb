```python
#Hypothesis testing:
import numpy as np
from scipy import stats
```

```python
data1=np.array([21,28,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
#perform t - test assuming unequal variances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results'
print("T-statistic:",t_stat)
print("P-value : ",p_value)
```

```python
#set siginificance level
alpha=0.05
```

```python
# Interpret the results
if p_value<alpha:
  print("NULL hypothesis no significance")
else:
  print("Alternate hypothesis significance")
```

```python
a=np.array(df.Efficiency)
b=np.array(df.Rec_Rate)
```

```python
alpha=0.05
```

ANOVA (ONE WAY testing)

```python
dfA=df[df['Vaccine']=='A']['Efficiency']
dfB=df[df['Vaccine']=='B']['Efficiency']
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print(f_stat)
print(p_value)
```

```python
#print the results
print("f-statistic:",f_stat)
print("P-value : ",p_value)
```

```python
#set siginificance level
alpha=0.05
```

```python
# Interpret the results
if p_value<alpha:
  print("NULL hypothesis no significance")
else:
  print("Alternate hypothesis significance")
```
